### Data Cleaning

```python
df["RID"].astype(str)
```

```python
df["RID"].replace(regex=True, inplace=True, to_replace=r'\D',value=r"")
```

```python
df["RID"]
```

#### Analysis 
Create New Year Column Based on Clean RIDs and Find Frequency of Year 

```python
df.reset_index(inplace=True)
```

```python
def year(RID):
    for i in RID:
        if RID.startswith("14"):
            return "2014"
        elif RID.startswith("15"):
            return "2015"
        elif RID.startswith("16"):
            return "2016"
        elif RID.startswith("17"):
            return "2017"
        elif RID.startswith("18"):
            return "2018"
        else: 
            return "NaN"
        
df["Year"] = df.RID.apply(year)
df
    
```

```python
df["RID"].astype(str)
```

```python
df["Year"].value_counts()
```

```python
bar = df["Year"].value_counts().plot('bar')
```
